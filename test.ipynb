```python
env = gym.make("ALE/Freeway-v5")

model = TRPO("MlpPolicy", env, gamma=0.99, verbose=1)
model.learn(total_timesteps=100_000, log_interval=4)
model.save("trpo_pendulum")

del model # remove to demonstrate saving and loading

model = TRPO.load("trpo_pendulum")

env = gym.make("ALE/Freeway-v5", render_mode="human")
obs = env.reset()
while True:
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, done, info = env.step(action)
    #env.render()
    #time.sleep(0.00001)
    print("uwu")
    if done:
        obs = env.reset()
```
